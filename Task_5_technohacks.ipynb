# TASK-5
Identify and remove duplicate values in an Iris dataset.

# Importing the python libraries

```python
import pandas as pd
```

# Importing the dataset

```python
df=pd.read_csv("Iris.csv")
df
```

```python
df.head(20)                                           #displaying first 20 rows of the dataset
```

# Checking duplicates of entire rows

```python
df.duplicated()
```

```python
df.duplicated().value_counts()
```

# Removing the rows which are completely duplicated

```python
df.drop_duplicates()
```

# Removing duplicates where the values are duplicated in one column

```python
df.drop_duplicates("Species")
```

```python
df.duplicated(subset=['SepalLengthCm']).value_counts()
```

# Removing duplicates based on more than one column

```python
df.drop_duplicates(["SepalLengthCm","SepalWidthCm"])
```

# Remove Duplicates based on more than one column and keep first/last

```python
df.drop_duplicates(["SepalLengthCm","SepalWidthCm"],keep="last")
```
